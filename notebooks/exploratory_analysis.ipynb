# Análise Exploratória do Dataset de CAPTCHAs
Este notebook explora o conjunto de dados de CAPTCHAs, mostrando estatísticas, exemplos e distribuições dos caracteres.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image
DATASET_TRAIN = '../dataset/train'
DATASET_TEST = '../dataset/test'
```

```python
def get_labels(dataset_path):
    files = [f for f in os.listdir(dataset_path) if f.endswith('.png')]
    labels = [os.path.splitext(f)[0] for f in files]
    return files, labels

train_files, train_labels = get_labels(DATASET_TRAIN)
test_files, test_labels = get_labels(DATASET_TEST)
print(f'Total de imagens de treino: {len(train_files)}')
print(f'Total de imagens de teste: {len(test_files)}')
```

```python
all_chars = ''.join(train_labels)
char_counts = Counter(all_chars)
plt.figure(figsize=(12,4))
plt.bar(char_counts.keys(), char_counts.values())
plt.title('Distribuição dos caracteres no treino')
plt.xlabel('Caractere')
plt.ylabel('Frequência')
plt.show()
```

```python
n = 10
plt.figure(figsize=(15,2))
for i, fname in enumerate(train_files[:n]):
    img = Image.open(os.path.join(DATASET_TRAIN, fname))
    plt.subplot(1, n, i+1)
    plt.imshow(img, cmap='gray')
    plt.title(os.path.splitext(fname)[0])
    plt.axis('off')
plt.suptitle('Exemplos de CAPTCHAs do treino')
plt.show()
```
